```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def scrape_zillow_data(url):
    chrome_options = Options()
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    
    driver = webdriver.Chrome(options=chrome_options)
    
    try:
        print("Navigating to Zillow...")
        driver.get(url)
        
        wait = WebDriverWait(driver, 15)
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '[data-test="property-card-price"]')))
        
        driver.execute_script("window.scrollTo(0, 500);")
        time.sleep(2) 
        
        print("Extracting data...\n")
        
        prices = driver.find_elements(By.CSS_SELECTOR, '[data-test="property-card-price"]')
        addresses = driver.find_elements(By.CSS_SELECTOR, '[data-test="property-card-addr"]')
        details = driver.find_elements(By.CSS_SELECTOR, '[data-test="property-card-details"]')
        
        for price, address, detail in zip(prices, addresses, details):
            print(f"Price:   {price.text}")
            print(f"Address: {address.text}")
            print(f"Details: {detail.text.replace('bds', 'beds').replace('ba', 'baths')}")
            print("-" * 40)
            
    except Exception as e:
        print(f"An error occurred: {e}")
        
    finally:
        driver.quit()

zillow_url = "https://www.zillow.com/homes/for_sale/?searchQueryState=%7B%22isMapVisible%22%3Afalse%2C%22mapBounds%22%3A%7B%22west%22%3A-125.87178595047308%2C%22east%22%3A-74.45577032547308%2C%22south%22%3A18.915001030362603%2C%22north%22%3A46.99403071678389%7D%2C%22filterState%22%3A%7B%22sort%22%3A%7B%22value%22%3A%22globalrelevanceex%22%7D%7D%2C%22isListVisible%22%3Atrue%2C%22curatedCollection%22%3Anull%2C%22mapZoom%22%3A5%7D"

scrape_zillow_data(zillow_url)
```
